```python
import os

def print_tree(path, prefix=""):
    # Print the current directory or file
    print(prefix + os.path.basename(path))

    if os.path.isdir(path):
        try:
            items = os.listdir(path)
        except PermissionError:
            print(prefix + "    [Permission Denied]")
            return

        for i, item in enumerate(items):
            full_path = os.path.join(path, item)

            # Decide prefix for children
            if i == len(items) - 1:
                new_prefix = prefix + "└── "
                child_prefix = prefix + "    "
            else:
                new_prefix = prefix + "├── "
                child_prefix = prefix + "│   "

            print_tree(full_path, child_prefix)

# ------------------ USE IT ------------------
folder_path = r"C:\Users\arnab\OneDrive\Desktop\Course Project\AIOS-LSFS_local"  # change this
print_tree(folder_path)
```
